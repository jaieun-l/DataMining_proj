# 데이터 불러오기

```python
path = '/content/gdrive/MyDrive/데마팀플/data/'

score_5 = ['eastar_5.csv', 'jeju_5.csv', 'jin_5.csv', 'tway_5.csv']
score_1 = ['eastar_1.csv', 'jeju_1.csv', 'jin_1.csv', 'tway_1.csv']

reviews_5 = pd.DataFrame()
reviews_1 = pd.DataFrame()

for i in range(len(score_5)):
    data = pd.read_csv(path + score_5[i])
    data = data.iloc[:,1]
    reviews_5 = pd.concat([reviews_5, data])
    
for i in range(len(score_1)):
    data = pd.read_csv(path + score_1[i])
    data = data.iloc[:,1]
    reviews_1 = pd.concat([reviews_1, data])
```

### 명사 추출 및 DTM 계산

```python
def NLP_DTM(review, num):  # 명사 추출 후 DTM 계산
    tagger = Okt()
    cv = CountVectorizer()
    corpus = []
    for doc_num in tqdm(range(5), desc='문서 진행도'):
        noun_list = tagger.nouns(review.iloc[doc_num,0])
        corpus.append(' '.join(noun_list))
    DTM_Array=cv.fit_transform(corpus).toarray()
    fname = cv.get_feature_names()
    DTMdf=pd.DataFrame(DTM_Array,columns=fname)
    DTMdf.to_csv(path+'DTM_reviews_'+str(num)+'.csv', encoding='utf=8-sig')

    return DTMdf
```

## 1점 리뷰

```python
reviews_1 = NLP_DTM(reviews_1, 1)

# 열별로 빈도 합 구해서 빈도 높은 순 출력
sum1 = reviews_1.sum(axis=0)

# 내림차순 술력
sum1 = sum1.sort_values(axis=0, ascending=False)
sum1
```

```python
# 1점짜리 리뷰 내 불용어
stopword = ['선택','때문','그냥','절대','다시','거기','다른','생각','중간','기타큐슈','권장','상황','시도','기준','각기','진에어','발생','이용','내내','다음','단위','부탁','달라','제주','김해','김포','당신','도대체','덕분','러시']

for i in stopword :
  if i in sum1 :
    sum1 = sum1.drop(i)

sum1 = sum1.sort_values(ascending=False).head(40)  # 단어 노출 40개로 제한
```

```python
cloud = WordCloud(font_path='/usr/share/fonts/truetype/nanum/NanumBarunGothic.ttf',background_color="white").generate_from_frequencies(sum1)

plt.figure(figsize=(14,10))
plt.imshow(cloud,interpolation='lanczos')
plt.axis('off')
plt.show()
```
