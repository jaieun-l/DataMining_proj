# Confusino Matrix, ROC 커브 시각화

```python
import pandas as pd
import numpy as np

import warnings
warnings.filterwarnings('ignore')

# 시각화
import seaborn as sn
import matplotlib.pyplot as plt

import graphviz
import sklearn.metrics as mt
from sklearn.tree import export_graphviz

# 모델
import statsmodels.api as sm

import xgboost as xgb
from xgboost import plot_importance


# 의사 결정 나무
from sklearn.tree import DecisionTreeClassifier

# 랜덤 포레스트
from sklearn.ensemble import RandomForestClassifier

# 부스팅
from sklearn.ensemble import AdaBoostClassifier
from sklearn.ensemble import GradientBoostingClassifier 
from xgboost import XGBClassifier


# SVM
from sklearn import svm
from sklearn.svm import SVC

# 로지스틱
from sklearn.feature_selection import SelectKBest, mutual_info_regression, SelectFromModel, f_regression
from sklearn.linear_model import LogisticRegression

# 정규화, 표준화
from sklearn.preprocessing import MinMaxScaler
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split

# 분류 모델 평가 지표
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, roc_auc_score, confusion_matrix
import sklearn.metrics as metrics
from sklearn.metrics import classification_report

from sklearn.model_selection import cross_val_score, cross_validate
```

## 데이터 불러오기

```python
path = 'data/'

def load_data(name) :
    data = pd.read_csv(path + name + '.csv')
    data = data.drop(columns='Unnamed: 0')
    return data
    
x_train = load_data('x_train')
x_test = load_data('x_test')
y_train = load_data('y_train')
y_test = load_data('y_test')
```

## 함수 선언

```python
def con_matrix(model, y_test, y_pred):

    # 분류정확도 
    accuracy = accuracy_score(y_test, y_pred)
    precision = precision_score(y_test, y_pred)
    recall = recall_score(y_test, y_pred)
    F1 = f1_score(y_test, y_pred)
    AUC = roc_auc_score(y_test, y_pred)

    # confusion matrix 
    con_mat = confusion_matrix(y_true=y_test, y_pred=y_pred) 
    
    # confusion matrix heatmap  
    plt.figure(figsize=(6,6)) 
    sn.heatmap(con_mat, annot=True, fmt=".3f", linewidths=.5, cmap = 'Blues_r', square = True)
    plt.ylabel('Actual label'); 
    plt.xlabel('Predicted label'); 
    all_sample_title = 'Accuracy Score: {:.4f}'.format(accuracy) 
    plt.title(all_sample_title, size = 14) 
    plt.show()
    
    print('Precision:{:.4f}'.format(precision))
    print('Recall: {:.4f}'.format(recall))
    print('F1: {:.4f}'.format(F1))
    print('AUC: {:.4f}'.format(AUC))
```

```python
def rocvis(true , prob , label, color ) :
    fpr, tpr, thresholds = metrics.roc_curve(true, prob)
    plt.plot(fpr, tpr, marker='.', label = label, color=color)
```

```python
x_train = pd.read_csv(path + 'x_train.csv', header=0)
x_train = x_train.drop('Unnamed: 0', axis=1)
x_test = pd.read_csv(path + 'x_test.csv', header=0)
x_test = x_test.drop('Unnamed: 0', axis=1)
y_train = pd.read_csv(path + 'y_train.csv', header=0)
y_train = y_train.drop('Unnamed: 0', axis=1)
y_test = pd.read_csv(path + 'y_test.csv', header=0)
y_test = y_test.drop('Unnamed: 0', axis=1)
```

# 의사결정나무

```python
x_train_dt = x_train
x_test_dt = x_test
y_train_dt = y_train
y_test_dt = y_test
```

```python
x_train_dt
```

```python
dt = DecisionTreeClassifier(max_depth=17, min_samples_split=19, random_state=156) # max_depth=17 으로 제한
dt.fit(x_train_dt, y_train_dt)
```

```python
y_pred_dt = dt.predict(x_test_dt)
con_matrix(dt, y_test_dt, y_pred_dt)
```

# 랜덤포레스트

```python
x_train_rf = x_train
x_test_rf = x_test
y_train_rf = y_train
y_test_rf = y_test
```

```python
# 랜덤 포레스트 하이퍼파라미터 튜닝 이후
rf = RandomForestClassifier(n_estimators = 25, 
                                max_depth = 12,
                                min_samples_leaf = 4,
                                min_samples_split = 4,
                                random_state = 0,
                                n_jobs = -1)
rf.fit(x_train_rf, y_train_rf)
y_pred_rf = rf.predict(x_test_rf)

con_matrix(rf, y_test_rf, y_pred_rf)
```

# Boosting

## Adaboost

```python
ada = RandomForestClassifier(n_estimators = 100, random_state = 0)
ada.fit(x_train_rf, y_train_rf)
y_pred_ada = ada.predict(x_test_rf)

con_matrix(ada, y_test_rf, y_pred_ada)
```

## GBM

```python
gbm = GradientBoostingClassifier(learning_rate=0.1,n_estimators=150,random_state=0)
gbm.fit(x_train_rf, y_train_rf.values)
y_pred_gbm = gbm.predict(x_test_rf)

con_matrix(gbm, y_test_rf, y_pred_gbm)
```

## Xgb

```python
# 모델 생성
xgb = XGBClassifier(learning_rate=0.1,max_depth=8,n_estimators=100)
xgb.fit(x_train_rf, y_train_rf) 
y_pred_xgb = xgb.predict(x_test_rf)

con_matrix(xgb, y_test_rf, y_pred_xgb)
```

# SVM

```python
train_data = pd.concat([x_train, y_train], axis=1)
test_data = pd.concat([x_test, y_test], axis=1)
test_data
```

```python
# 데이터 축소
train = train_data.sample(frac=0.08)
test = test_data.sample(frac=0.08)

y_train_svm = train['satisfaction']
y_test_svm = test['satisfaction']

x_train_svm = train.drop(columns='satisfaction')
x_test_svm = test.drop(columns='satisfaction')

x_test_svm
```

```python
# 표준화

scaler = StandardScaler()
x_train_standard = scaler.fit_transform(x_train_svm)
x_test_standard = scaler.transform(x_test_svm)
x_train_standard
```

```python
svc = SVC(kernel='rbf', C=1, gamma=0.1, probability=True)
svc.fit(X=x_train_standard, y=y_train_svm)
```

```python
y_pred_svm = svc.predict(x_test_standard)
y_pred_svm = pd.DataFrame(y_pred_svm)

con_matrix(svc, y_test_svm, y_pred_svm)
```

# 로지스틱 회귀

```python
x_train_lr = x_train
x_test_lr = x_test
y_train_lr = y_train
y_test_lr = y_test
```

```python
def logisticModel(x_train_lr, x_test_lr, y_train_lr, y_test_lr) :
    log_reg = LogisticRegression(random_state=0, multi_class='multinomial')

    log_reg.fit(X=x_train_lr, y=y_train_lr)

    y_pred_lr = log_reg.predict(X=x_test_lr)
    y_pred_lr = pd.DataFrame(y_pred_lr)
    con_matrix(log_reg, y_test_lr, y_pred_lr)

def logitModel(x_train_lr, x_test_lr, y_train_lr, y_test_lr) :
    log_reg = sm.Logit(y_train_lr, x_train_lr)
    results = log_reg.fit(method = "newton")

    print(np.exp(results.params), results.summary2())
```

```python
# 지연 시간 변수 생성 (all_delay_in_minutes : 도착-출발)
x_train_logit = x_train_lr.copy()
x_test_logit = x_test_lr.copy()

x_train_logit.insert(len(x_train_lr.columns) - 1,'all_delay_in_minutes', x_train_lr['arrival_delay_in_minutes'] - x_train_lr['departure_delay_in_minutes'])
x_test_logit.insert(len(x_test_lr.columns) - 1,'all_delay_in_minutes', x_test_lr['arrival_delay_in_minutes'] - x_test_lr['departure_delay_in_minutes'])

# 출발, 도착 변수 삭제
x_train_logit = x_train_logit.drop(['arrival_delay_in_minutes', 'departure_delay_in_minutes'], axis = 1)
x_test_logit = x_test_logit.drop(['arrival_delay_in_minutes', 'departure_delay_in_minutes'], axis = 1)

x_train_logit
```

```python
# 표준화
scaler = StandardScaler()
scaler.fit(x_train_logit)
x_trains = scaler.transform(x_train_logit)
x_tests = scaler.transform(x_test_logit)

# 정규화
scaler = MinMaxScaler()
scaler.fit(x_trains)
x_trainsn = scaler.transform(x_trains)
x_testsn = scaler.transform(x_tests)
x_trainsn
```

```python
# 변수 선택

# target(satisfaction)와 가장 correlated 된 features 를 k개 고르기..
predict = pd.DataFrame(columns = ['train', 'test'])
train_list = []
test_list = []

for i in range(1, 22) :
    selector = SelectKBest(score_func=f_regression, k=i) ## selctor 정의하기

    x_train_selected = selector.fit_transform(x_trainsn, y_train_lr) ## 학습데이터에 fit_transform 
    x_test_selected = selector.transform(x_testsn) ## 테스트 데이터는 transform

    model = LogisticRegression()
    model.fit(x_train_selected, y_train_lr)
    
    train_list.append(model.score(x_train_selected, y_train_lr))
    test_list.append(model.score(x_test_selected, y_test_lr))
    
predict['train'] = train_list
predict['test'] = test_list
predict
```

```python
# 예측 정확도 가장 높은 변수 개수 탐색
if train_list.index(max(train_list)) == test_list.index(max(test_list)) :
    optIndex = train_list.index(max(train_list))
    print(optIndex, max(train_list), max(test_list))
else :
    optIndex = int((train_list.index(max(train_list)) + test_list.index(max(test_list))) / 2)
    print(optIndex, max(train_list), max(test_list))
```

```python
# target(satisfaction)와 가장 correlated 된 features 개수 선택

selector = SelectKBest(score_func=f_regression, k=optIndex) ## selctor 정의하기

x_train_selected = selector.fit_transform(x_train_lr, y_train_lr) ## 학습데이터에 fit_transform 
x_test_selected = selector.transform(x_test_lr) ## 테스트 데이터는 transform

x_train_selected.shape, x_test_selected.shape
```

```python
# 학습결과 평가

log_reg = LogisticRegression(random_state=0, multi_class='multinomial')

log_reg.fit(X=x_train_selected, y=y_train)

y_pred_lr = log_reg.predict(X=x_test_selected)
con_matrix(log_reg, y_test_lr, y_pred_lr)
```
